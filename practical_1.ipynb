```python
FACTORIAL :
```

```python
def factorial(n):

    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

number = 5
print(f"The factorial of {number} is {factorial(number)}")

```

```python
MULTIPLICATION :
```

```python
def display_multiplication_table(number, limit=10):

    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")


num = 7
display_multiplication_table(num)

print("\n")
num_2 = 12
limit_2 = 5
display_multiplication_table(num_2, limit_2)
```

```python
EVEN AND ODD :
```

```python
def check_odd_even(number):

    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Test with various numbers
print(check_odd_even(4))
print(check_odd_even(7))
print(check_odd_even(0))
print(check_odd_even(15))
```

```python
PALINDROME :
```

```python
def is_palindrome(number):
    """Checks if a given number is a palindrome."""

    num_str = str(number)

    reversed_num_str = num_str[::-1]

    if num_str == reversed_num_str:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."


print(is_palindrome(121))
print(is_palindrome(123))
print(is_palindrome(1331))
print(is_palindrome(7))
print(is_palindrome(12345))
```

```python
EXPERT SYSTEM :
```

```python
def expert_system():
    print("=== Simple Medical Expert System ===")

    fever = input("Do you have fever? (yes/no): ").lower()
    cough = input("Do you have cough? (yes/no): ").lower()
    headache = input("Do you have headache? (yes/no): ").lower()

    # Rule-based decision making
    if fever == "yes" and cough == "yes" and headache == "yes":
        print("Diagnosis: You may have Flu.")
    elif fever == "yes" and cough == "yes":
        print("Diagnosis: You may have Common Cold.")
    elif headache == "yes":
        print("Diagnosis: You may be stressed or dehydrated.")
    else:
        print("Diagnosis: No serious illness detected.")

# Run the expert system
expert_system()
```
